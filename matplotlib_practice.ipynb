```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20 ,35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
# grid
plt.grid()
```

```python
# Skill Drill
fig, ax = plt.subplots()
# add color and legend (c is short hand for cyan per documentation)
ax.barh(x_axis, y_axis, color="c", label='Chicago')
# invert months descending
ax.invert_yaxis()
# add a title
ax.set_title("PyBer Fare by Month")
# add axes labels
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# show the legend
ax.legend()
```

```python
# Skill Drill
plt.scatter(y_axis, x_axis, color='r', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date)')
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# make the points larger with s (size) dependent on y_axis
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# instead of creating variable and for loop, use list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill Drill
fig, ax = plt.subplots()
# alpha is transparency and edgecolors adds a color to the circles
ax.scatter(y_axis, x_axis, color='skyblue', alpha=0.8, edgecolors='black', s = [i * 5 for i in y_axis]
          , label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```

```python
# Skill Drill
colors = ["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
shadow=(0.8)
fig, ax = plt.subplots()
ax.pie(y_axis,
      labels=x_axis,
      colors=colors,
      explode=explode_values,
        shadow=shadow,
       startangle=95,
       counterclock=False,
      autopct='%.1f%%')
plt.show()
```
